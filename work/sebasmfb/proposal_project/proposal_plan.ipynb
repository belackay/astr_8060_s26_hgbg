## Scientific Justification

We expect Supermassive Black Hole Binaries as natural consequences

We expect supermassive black hole binaries (SMBHBs) are natural consequences of galaxy mergers.

A series of dissipative interactions bring SMBHs in merging galaxies toward a common merger remnant. Observations of closely-separated dual active galactic nuclei (AGN) (~7 pc) anchor of the early frames of this picture.

Irregularities in AGN time-series spectra such as two-peakedness and a highly shifted broad line can indicate bulk orbital motion in a binary, as certain regions of the AGN spectra will be subject to Doppler boosting.

Time-domain photometric observations also may reveal orbital motion via periodic variability. Particularly, since the aforementioned analysis is most sensitive to binaries with periods on the order of 5-20 years, we plan to take long-term photometric observations at a consistent cadence for combination with archival data of spectroscopically-identified SMBHB candidates in order to support or refute their status as a binary. We turn to Banaszak et al. 2026 for a joint model selection scheme which is robust against False Positives, and we use the same source to discern which observing pattern is ideal for our use case.

#### Steps:
- Use catalogs of spectroscopically-identified binary AGN candidates (current work at Carnegie Mellon University).
- Compile archival light curves for said candidates, use joint model selection outlined in Banasazk et al. 2026 to idenify standout candidates for follow-up observations
- Take and append photometric observations with Las Cumbres Observatory's 1-meter telescope to these light curves
- Analyze light curve with the same joint model selection procedure.


## Technical Justification

DESI and Las Cumbres cover an overlapping region of sky, so cna be used in conjunction.

I haven't chosen the targets YET, but I will choose ones which are near or below the r-band 5-sigma depth of ~21.5 - 22.0 AB mags (there will likely be quite a few targets to choose from). This 5-sigma depth is shallow compared to LSST, but it is the best available for an instrument which focuses on time-domain optical photometry.

#### Instrument:

Las Cumbres Observatory's 1-meter telescope with the Sinistro imager in SDSS r-band


#### Exposure Time:

I would use the 5-sigma depth of the LCO and the AB magnitudes of my target to establish the relationship between exposure time and photometric uncertainty - typically, the LCO uses exposure times of 300 - 600 seconds for observations with a magnitude near the 5-sigma depth. To determine which exposure time to use, I will roughly calculate the minimum photometric uncertainty we want, generally based on the analysis in Banaszak et al. 2026 which distinguishes performance between imprecise data from CRTS and precise data from ZTF and LSST.
