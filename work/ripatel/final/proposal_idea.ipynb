### Scientific Justification:
I want to find recoiling black holes. We know they have to exist because we have detected mergers from black hole binaries. We also know from numerical relativity simulations that asymmetries in the mass ratios or black hole spin configurations should produce a recoil, or "kick", which causes black hole remnants to be off-nuclear. We have many candidates for a recoiling black hole, but we have not been able to view one yet. 

Even the best recoiling AGN candidates (e.g., CID-42) have recently been ruled out with high-resolution imaging, highlighting the need for targeted, high-precision spatial studies.

#### Abstract (in progress):
A black hole merger remnant can acquire a recoil velocity due to anisotropic gravitational wave emission, which can cause it to be displaced from the center of its galaxy. Although spectroscopic surveys have found many recoiling active galactic nuclei (AGN) candidates by identifying offset broad H-beta emission lines, none have been confirmed through direct photometric imaging. We propose near-infrared (NIR) imaging of a selected subsample of these recoiling AGN candidates. We have selected these candidates based on existing radio detections, and having well-characterized host galaxies. Our goal is to measure the offset between the AGN and the photometric center of the host galaxy. We will model the host galaxy light distribution using parametric fitting with GALFIT to determine its center, and compare this to the AGN position constrained by high-resolution radio observations. A statistically significant spatial offset between the AGN and host galaxy center would provide direct evidence for a recoiling black hole. Confirming such systems would allow us to improve our understanding of AGN kinematics, constrain black hole merger dynamics, provide observational inputs for future gravitational wave missions such as LISA and pulsar timing arrays. 
Non-detections will place quantitative limits on the reliability of spectroscopic selection techniques and improve our understanding of AGN kinematics and host galaxy structure.
