### Sample Selection

Have not made a sample yet but here is my conceptualized framework for choosing candidates:

Base catalog: SDSS DR17 IMBH AGN sample (930 broad-line IMBH AGNs).


- Apply selection cuts:

    - Brightness cut: r ≤ 21.5 (lenient) or r ≤ 21 (conservative)
        - Why this cut?
            - Use as a starting parent-sample cut, but I will be refined with the DECam ETC to guarantee high S/N.
            - Look at previous paper with DECam to look for previous magnitudes
        - Will have to query r-band magnitude 


    - Redshift: z ≤ 0.2 (or whichever maximizes angular size/host light control).
        - Keeping objects local will ensure their angular sizes are large enough to keep the physical size to angular size ratio high enough to have the galaxy cover multiple pixels (since DECam has 0.2637 arcsec/pixel). At z=0.2, a pixel will cover a physical length of 0.87 kpc (see cell below for calulation)
            - Having small physical size to pixel scales allows better isolation from host galaxy (mimizing contamination from host)


    - Remove targets outside accessible RA/Dec (if any)


Final sample size: hopefully several targets for the primary experiment 


```python
import numpy as np
from astropy.cosmology import FlatLambdaCDM


cosmo = FlatLambdaCDM(H0=70, Om0=0.3)
z = 0.20

# tan(x) = Distance / Physical Size --> small angle approx --> x ~ Distance/physical size --> Physical size = x * Distance
DA = cosmo.angular_diameter_distance(z).value  # angular diameter distance [Mpc]
theta= np.pi/180* 1/(3600) # arcseconds in 1 degree

scale_kpc_per_arcsec = theta * DA*1000 #multiplied DA by 1000 to get to kpc scales
print(scale_kpc_per_arcsec, "kpc/arcsec")

pixel_scale= 0.2637 #arcsec/pix

print((pixel_scale* scale_kpc_per_arcsec),"kpc/pixel")
```

### Observing Strategy (instrument + exposure plan)

$\textbf{Filters:}$

    - Will take exposures in both filters for sure!

- Exposure times:
