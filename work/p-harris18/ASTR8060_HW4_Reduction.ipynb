## Bias Variation

Examining how detector bias level changes during the night by computing mean pixel value in a central region of each frame. Plotting the resulting values shows the stability of the bias level

```python
import os
import numpy as np
from astropy.io import fits

bias_files = sorted([f for f in os.listdir() if f.startswith("d") and f.endswith(".fits")])
print("Bias files:", bias_files)

bias_means = []
bias_rms = []

for f in bias_files:
    with fits.open(f) as hdul:
        data = hdul[0].data
        
        region = data[900:1100, 900:1100]
        
        bias_means.append(np.mean(region))
        bias_rms.append(np.std(region))

print("Mean bias levels:", bias_means)
print("RMS values:", bias_rms)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(len(bias_files)), bias_means, marker='o')
plt.xlabel("Bias frame index")
plt.ylabel("Mean counts (ADU)")
plt.title("Bias level variation through the night")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(range(len(bias_files)), bias_rms, marker='o', color='orange')
plt.xlabel("Bias frame index")
plt.ylabel("RMS (ADU)")
plt.title("Bias RMS variation through the night")
plt.grid(True)
plt.show()
```

### Review of Bias Variation

The mean bias level changes slightly over the course of the night, decreasing from approx. 666 ADU to about 661 ADU. This indicates small scale drifting is occurring in the detector bias level during observation.

The RMS values are relatively constant at roughly 4.5 ADU, meaning the read noise of the detector is stable.

The bias level varies by only a few ADU with stable detector noise, therefore it is reasonable to combine all of the biases into a master bias. 

## Overscan Subtraction and Image Trimming

Using CCDPROC.SUBTRACT_OVERSCAN and CCDPROC.TRIM_IMAGE to fit and subtract the overscan region of each image, then trim the image to remove the overscan region

```python
import ccdproc
from astropy.nddata import CCDData
import astropy.units as u

hdul = fits.open("a004.fits")
data = hdul[0].data

ccd = CCDData(data, unit="adu")

ccd_overscan_sub = ccdproc.subtract_overscan(ccd, overscan=ccd[:, 2050:])

ccd_trimmed = ccdproc.trim_image(ccd_overscan_sub[:, :2048])
```

```python
plt.imshow(ccd_trimmed.data, cmap="gray", origin="lower")
plt.colorbar()
plt.title("Trimmed CCD Image")
plt.show()
```

## trying out different overscan fitting functions

Testing legendre, hermite, and polynomial in order of 2

```python
from astropy.modeling import models

leg2 = ccdproc.subtract_overscan(
    ccd,
    overscan=ccd[:,2050:],
    model=models.Legendre1D(2)
)

herm2 = ccdproc.subtract_overscan(
    ccd,
    overscan=ccd[:,2050:],
    model=models.Hermite1D(2)
)

poly2 = ccdproc.subtract_overscan(
    ccd,
    overscan=ccd[:,2050:],
    model=models.Polynomial1D(2)
)
```

```python
plt.figure()
plt.imshow(leg2.data, cmap="gray", origin="lower")
plt.title("Overscan subtraction – Legendre (order 2)")
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(herm2.data, cmap="gray", origin="lower")
plt.title("Overscan subtraction – Hermite (order 2)")
plt.colorbar()
plt.show()

plt.figure()
plt.imshow(poly2.data, cmap="gray", origin="lower")
plt.title("Overscan subtraction – Polynomial (order 2)")
plt.colorbar()
plt.show()
```

```python
from astropy.modeling import models

orders = range(1,9)

for order in orders:
    model = models.Chebyshev1D(order)

    corrected = ccdproc.subtract_overscan(
        ccd,
        overscan=ccd[:,2050:],
        model=model
    )

    plt.figure()
    plt.imshow(corrected.data, cmap="gray", origin="lower")
    plt.title(f"Chebyshev fit order {order}")
    plt.colorbar()
    plt.show()
```

## Review of overscan subtraction and trimming

the different overscan fitting functions are giving similar overscan subtraction results. Testing chebyshev fits from order 1 through 8 produce nearly identical images and shows that increasing the order by more than a few does not significantly improve the results. It may begin to fit more noise instead. Orders 2 to 4 appear to be sufficient for accurately modeling these data.

## Master Bias Frame

using CCDPROC.COMBINE to combine the bias frames into a master bias frame and evaluating trends.

```python
from astropy.nddata import CCDData
import astropy.units as u
import ccdproc

bias_ccds = []
for f in bias_files:
    ccd_bias = CCDData.read(f, unit="adu")
    bias_ccds.append(ccd_bias)

master_bias = ccdproc.combine(bias_ccds, method='median')

print(master_bias)
print(master_bias.shape)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(master_bias.data, cmap="gray", origin="lower")
plt.colorbar()
plt.title("Master Bias Frame")
plt.show()
```

```python
master_region = master_bias.data[900:1100, 900:1100]

master_mean = np.mean(master_region)
master_rms = np.std(master_region)

print("Master bias mean:", master_mean)
print("Master bias RMS:", master_rms)
```

## Evaluation of trends

The RMS of about 1.5 ADU is much lower than that of the individual bias frames with an RMS of roughly 4.5 ADU. Combining multiple bias frames reduces a significant amount of noise. A change in the bias level of approximately 5 ADU occurs over the observational period. The pattern of the individual bias frames is consistent, indicating that the detector is remaining stable and that the variation is due to a small shift in the bias level.

Doing a bias subtraction in addition to the overscan subtraction would be adding noise equal to the RMS of the master bias, so it would add about 1.5 ADU of noise.

```python
dark_files = sorted([f for f in os.listdir() if f.startswith("d") and f.endswith(".fits")])

dark_currents = []

for f in dark_files:
    dark = CCDData.read(f, unit="adu")
    dark_sub = dark.subtract(master_bias)

    region = dark_sub.data[900:1100, 900:1100]
    dark_mean = np.mean(region)

    dark_currents.append(dark_mean)

print("Mean dark signal (ADU):", np.mean(dark_currents))
```

```python
gain = 2.5   # electrons per ADU
exptime = 2  # seconds

dark_rate = np.mean(dark_currents) * gain / exptime
print("Dark current (e-/pixel/sec):", dark_rate)
```

## Dark current

The measured mean signal is approx. -0.036 ADU after subtracting the master bias from the dark frames. Dark current was determined to be approx. -0.045 electrons/pixel/sec after converting the measured mean signal into electrons using the detector gain of 2.5 electrons/ADU, and then dividing by the exposure time of 2 seconds. The nearly zero result indicates that the dark current measured in the detector is negligible and does not significantly contribute to noise.

```python
from astropy.nddata import CCDData
import ccdproc
import os

def process_frame(filename, master_bias):
    ccd = CCDData.read(filename, unit="adu")
    
    ccd_os = ccdproc.subtract_overscan(ccd, overscan=ccd[:, 2050:])
    
    ccd_trim = ccdproc.trim_image(ccd_os[:, :2048])

    master_bias_trim = ccdproc.trim_image(master_bias[:, :2048])
    
    ccd_cal = ccd_trim.subtract(master_bias_trim)
    
    return ccd_cal
```
